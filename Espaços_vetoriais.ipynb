**Espaço vetorial: definição**

*Um [espaço vetorial](https://pt.wikipedia.org/wiki/Espa%C3%A7o_vetorial), sobre o [corpo $K$](https://pt.wikipedia.org/wiki/Corpo_(matem%C3%A1tica)), é um conjunto $V$, não vazio, cujos elementos possuem as seguintes propriedades:*

1.   Se $u$ e $v$ forem elementos de $V$ é possível "somar" esses elementos e obter um elemento $w=u+v$ que ainda é um elemento de $V$.
2.   Se $u$ for um elemento de $V$ e $\alpha$ for um elemento do corpo $K$, é possível "multiplicar" esses elementos e obter um elemento $\alpha u$ que ainda é um elemento de $V$.

Exigimos ainda que, se $u$, $v$ e $w$ forem elemementos de $V$ e $\alpha$ e $\beta$ forem elementos de $K$, então:


1. $u+(v+w)=(u+v)+w$; (<font color =blue> associativa) <font color =black>

2. $u+v=v+u$; (<font color =blue> comutativa) <font color =black>

3. Existe um elemento $0 \in  V$, denominado vetor nulo, tal que $v + 0 = v$; 

4. 	Para todo $v \in V$, existe um elemento $−v \in V$, chamado de inverso aditivo de $v$, tal que $v + (−v) = 0$; <font color=blue> (simétrico)<font color =black>
5. $\alpha (u + v) = \alpha u + \alpha v$; (<font color =blue> distributiva)<font color =black>

6.   <font color =black>
 $(\alpha\beta)u = \alpha(\beta u)$; (<font color =blue> associativa)<font  color =black>

7. $(\alpha + \beta)v = \alpha v + \beta v$; (<font color =blue> distributiva)<font color =black>

8. $1v = v$, em que 1 denota a identidade multiplicativa em $K$.

#**Um exemplo: sobre vetores, tabelas e matrizes**

**Obs:** Em Python, a primeira coordenada do vetor é a 0.

Uma empresa produz três produtos (A, B e C) e suas despesas de produção são divididas em três categorias (Matérias-primas, Mão de obra e Outras despesas). Veja as tabelas a seguir, com os custos de produção:

```python
Despesas=['Matérias-primas','Mão de obra','Outras despesas']   # Lista de nomes, strings.
Produto=['A','B','C']
Dadosproducao=[[0.1,0.3,0.15],[0.3,0.4,0.25],[0.1,0.2,0.15]]   # Isso é uma lista, com três listas de números.

Dadosproducao=np.array(Dadosproducao)                          # Agora é uma matriz, com três linhas.           

import pandas as pd                                            # Para lidar com tabelas ou DataFrame.
df = pd.DataFrame(Dadosproducao,index=Despesas,columns=Produto)
print(df)

print()
df1 = pd.DataFrame(np.transpose(Dadosproducao),index=Produto,columns=Despesas)
print(df1)
print()                     # Mostra índices de linhas e colunas.
```

Você pode fazer gráficos com tabelas, como no Excel.

```python
import matplotlib.pyplot as plt                    # Para fazer gráficos.

plt.plot(df)
plt.title('Gráfico de tabela de custos de produção por item.')
plt.show()

df1.plot(title='Custos de produção por Item (dólares)')
plt.show()
```

As quantidades produzidas nas estações do ano, ou nos trimestres, (Verão, Outono, Inverno e Primavera), são dadas na tabela a seguir:

```python
Estacao=['Verão','Outono','Inverno','Primavera']
Dadosquandidade=[[4000,4500,4500,4000],[2000,2600,2400,2200],[5800,6200,6000,6000]]
Dadosquandidade=np.array(Dadosquandidade)

df2 = pd.DataFrame(Dadosquandidade,index=Produto,columns=Estacao)
print(df2)
print()
df3 = pd.DataFrame(np.transpose(Dadosquandidade),index=Estacao,columns=Produto)
print(df3)
print()
df3.plot(title='Quantidade Produzida por Trimestre')
plt.legend(bbox_to_anchor =(1, 0.8), ncol = 2)        # Escolha de onde colocar a legenda.
plt.show()
```

Para uma reunião com os acionistas é desejado saber o custo total da produção, para cada categoria (Matérias-primas, Mão de obra, Outras despesas) e em cada trimestre (Verão, Outono, Inverno, Primavera).

Para obtermos esses valores, basta multiplicarmos as matrizes que representam as tabelas. No caso, temos duas matrizes $M$ e $P$, calculamos a matriz $S=MP$, que dos dá os custos.

Relembrando as tabelas:

```python
print(df)
print()

print(df2)
print()  
```

Isso nos diz que as matrizes $M$ e $P$ são:

```python
M=np.array(Dadosproducao)
P=np.array(Dadosquandidade)
print(f'Matriz M=')
print(M)
print()
print(f'Matriz P=')
print(P)
```

Observando a linha Matérias-primas vemos que o custo com matéria prima (no verão) é dado pela soma dos custos dos produtos $A$, $B$ e $C$ pela quantidade de cada produto $A$, $B$ e $C$ (no verão). No caso, é 
$$\begin{bmatrix}0.1&0.3&0.15\end{bmatrix}\begin{bmatrix}4000\\2000\\5800\end{bmatrix}=400+600+870=1870,\tag{$\star$}$$ que é o produto da primeira linha de $M$ com a primeira coluna de $P$.

Fazendo raciocínio semelhante com cada linha de $M$ e cada coluna de $P$ teremos:

```python
M=np.array(Dadosproducao)
P=np.array(Dadosquandidade)
Solucao=np.dot(M,P)                   # Calcular o produto escalar de cada linha de M, por cada coluna de P.

df4 = pd.DataFrame(Solucao,index=Despesas,columns=Estacao)

print(f'Tabela com custos')
print(df4)
```

Note que poderia calcular o produto de matrizes como $M@P$.

```python
M@P
```

Para obtermos agora o custo total (de toda a produção) por trimestre, basta somar os elementos de cada coluna da última matriz $S$.

```python
Solucaot=np.transpose(Solucao)
CustoTotal=[sum(Solucaot[0]),sum(Solucaot[1]),sum(Solucaot[2]),sum(Solucaot[3])]
print(f'Custo total de produção, por trimestre é {CustoTotal}')
```

Podemos juntar os cálculos em uma matriz:

```python
import copy as cp                    #Para copiar coisas, de forma que as alterações não interfiram no original.

New_Solucao=cp.deepcopy(Solucao.tolist()) # .tolist() volta para lista para incluir uma linha com o .append().
New_Solucao.append(CustoTotal)
New_Solucao=np.array(New_Solucao)

df5 = pd.DataFrame(New_Solucao,index=Despesas+['Custo total por trimestre'],columns=Estacao)

print(f'Tabela com custos')
print(df5)

df5.plot(title='Custos de produção')
plt.legend(bbox_to_anchor =(1, 0.8), ncol = 2) 
plt.show()
```

Se quisermos saber o custo total da produção no ano basta somar agora os elementos de cada linha da última matriz.

#**Sobre a notação de matrizes**

De forma geral teremos (veja esse [link](https://pt.wikipedia.org/wiki/Matriz_(matem%C3%A1tica)) para mais detalhes sobre matrizes)
$$M=\begin{bmatrix}m_{11}& m_{12}& \ldots&m_{1n}\\m_{21}& m_{22}& \ldots&m_{2n}\\\vdots& \vdots& \ddots&\vdots\\m_{p1}& m_{p2}& \ldots&m_{pn}\end{bmatrix},$$ como a representação de uma matriz, com $p$ linhas e $n$ colunas. Na forma resumida denotamos também 
$$M_{p\times n}=[m_{ij}]_{p\times n}.$$ 
Denotamos ainda por $$M^t=\begin{bmatrix}m_{11}& m_{21}& \ldots&m_{p1}\\m_{12}& m_{22}& \ldots&m_{p2}\\\vdots& \vdots& \ddots&\vdots\\m_{1n}& m_{2n}& \ldots&m_{pn}\end{bmatrix},$$ em que as linhas de $M$ tornam-se colunas de $M^t$.

É claro que podemos denotar uma matriz usando outras letras, como $A$, $B$,..., ou com índices $M_1$, $M_2$,..., $M_q$.

A soma usual de matrizes (a que vamos usar sempre nessa disciplina) $A_{p\times n}=[a_{ij}]_{p\times n}$ e $B_{p\times n}=[b_{ij}]_{p\times n}$ é dada por 
$$A+B=[a_{ij}+b_{ij}]_{p\times n}.$$ 

O produto usual das matrizes (o que vamos usar sempre nessa disciplina) $C_{p\times n}=[c_{ij}]_{p\times n}$ e $D_{n\times q}=[d_{ij}]_{n\times q}$ é dado por  
$$CD=\left[\sum_{k=1}^nc_{ik}d_{kj}\right]_{p\times q}.$$

Cabe observar que $$\sum_{k=1}^nc_{ik}d_{kj}=\begin{bmatrix}c_{i1}&c_{i2}&\cdots&c_{in}\end{bmatrix}\begin{bmatrix}d_{1j}\\d_{2j}\\\vdots&\\d_{nj}\end{bmatrix}=\begin{bmatrix}d_{i1}&d_{i2}&\cdots&d_{in}\end{bmatrix}\begin{bmatrix}c_{1j}\\c_{2j}\\\vdots&\\c_{nj}\end{bmatrix}$$ é o produto escalar da $i$-ésima linha de $A$ e a $j$-ésima coluna de $B$.

Você pode verificar que $$B^tA^t=(AB)^t.$$

<font color=blue> Uma outra observação importante para o que queremos nessa disciplina é que, quando denotamos os vetores (como colunas)
$$b_j=\begin{bmatrix}b_{1j}\\b_{2j}\\\vdots&\\b_{nj}\end{bmatrix},$$ dizemos que a Expressão ($\star\star$)
$$\sum_{j=1}^p\alpha_{j}b_i=\begin{bmatrix}b_{11}& b_{12}& \ldots&b_{1p}\\b_{21}& m_{22}& \ldots&m_{2p}\\\vdots& \vdots& \ddots&\vdots\\m_{n1}& m_{n2}& \ldots&m_{np}\end{bmatrix}\begin{bmatrix}\alpha_{1}\\\alpha_{2}\\\vdots\\\alpha_{p}\end{bmatrix}=B\alpha.\tag{$\star\star$}$$ 
 é uma combinação linear dos vetores $b_1,\,b_2,\,\ldots$ e $b_p$. 

<font color=blue> Se escrevermos os vetores $b_i$ como linhas, então teremos 
$$\sum_{j=1}^p\alpha_{j}b_i=\alpha^tB^t\tag{$\star\star^*$}$$ como a combinação linear dos mesmos vetores.

#**Exemplo 3:**

Um exemplo de espaço vetorial sobre o conjunto dos números reais $\mathbb{R}$ é o conjunto $$V=\{p(s)=a_2s^2+a_1s+a_0:\, a_2,\,a_1,\,a_0\in \mathbb{R},\quad s\in \mathbb{R}\}, $$ que é o conjunto das funções polinomiais, de grau no máximo $2$, e com coeficientes reais.

Observe que esse conjunto pode ser identificado com o espaço vetorial $$W=\{(a_2,a_1,a_0):\, a_2,\,a_1,\,a_0\in \mathbb{R}\}=\mathbb{R}^3.$$

Veja abaixo um exemplo de possível forma de identificar cada vetor $v\in \mathbb{R}^3$ com um poinômio $p_v(s)\in V$:

```python
def poli(v,s):
  p=0
  for i in range(0,len(v)):
    p=p+v[i]*s**(len(v)-i-1)
  return p
```

Para cada vetor $v$ fixo, temos uma função polinomial associada:

```python
v=[1,2,-4]

def p_v(s):  # Definição de p(s)=s^2+2s+4, por meio do vetor dos coeficentes v=[1,2,-4]
    return poli(v,s) 

u=[-1,6,-4]

def p_u(s):  # Definição de p(s)=-s^2+6s+4, por meio do vetor dos coeficentes u=[-1,6,-4]
    return poli(u,s) 

def p_umaisv(s):  # Definição de p_u(s)+2p_v(s)
    return poli(u,s)+2*poli(v,s)  
```

Note que $v=[1,2,-4]$, por exemplo, é um vetor (no sentido de Geometria Analítica) e $p_v(s)=s^2+2s-4$ é uma função, que também vemos como um vetor.

```python
x=np.linspace(-10,10,1000)                # Cria lista de 1000 númemros entre -10 e 10, inclusive, igualmante espaçados
y=[p_v(s) for s in x]                     # Calcula pv(s) para cada s nalista anterior.

plt.plot(x,y)                             # Faz gráfico.
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de p_v(s)')
plt.show()

z=[p_u(s) for s in x]
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de p_u(s)')
plt.show()

w=[p_umaisv(s) for s in x]
plt.plot(x,w)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de p_u(s)+2p_v(s)')
plt.show()
```

#**Combinações lineares de elementos de espaços vetoriais**

Como já mencionado anteriormente, um dos grandes interesses na disciplina de Álegbra Linear é em expressões da forma
$$w=\alpha_1u_1+\alpha_2u_2+\cdots+\alpha_nu_n,\tag{combinação linear}$$ em que 
$\alpha_1,\,\alpha_2,\,\ldots,\,\alpha_n\in \mathbb{R}$ e $u_1,\,u_2,\,\ldots,\,u_n\in V$, sendo $V$ um espaço vetorial, com as operações de "adição" $u+v$ e "multiplicação por escalar" $\alpha u$, para $u,v\in V$ e $\alpha\in \mathbb{K}$. 

Vamos chamar o vetor $w$, definido acima, de <font color=blue> **combinação linear** <font color=black>  dos vetores $u_1,\,u_2,\,\ldots,\,u_n\in V$.

**Exemplo 1:** Tomando $V=\mathbb{R}^2$ e $u=(1,2)$ então o conjunto 
$$U=\{\alpha u=\alpha(1,2):\, \alpha\in \mathbb{R}\}$$ é o conjunto das combinações lineares de $u$. Podemos verificar que $U$ é também um espaço vetorial, que está contido em $V$.

```python
x=np.linspace(-5,5,1000) # cria uma lista com 1000 valores entre -5 e 5, igualmente espaçados.
y=2*x

plt.plot(x,y,color='blue')
plt.title('Representação do conjunto U')
plt.show()
```

**Exemplo 2:** Seja  $$V = \{(x,y) \in \mathbb{R}^2: x,y > 0 \},$$ o espaço vetorial, com as operações de "adição"
$$\oplus:Q\times Q \rightarrow Q$$
$$(x_1,y_1)\oplus(x_2,y_2)=(x_1x_2,y_1y_2),$$
 e "multiplicação por escalar" dada por 
$$\odot : \mathbb{R} \times Q \rightarrow Q$$
$$c\odot(x,y)=(x^c,y^c).$$ 

Tomando $u=(2,3)\in V$, então o conjunto
$$U=\{\alpha u=(2^{\alpha},3^{\alpha}):\, \alpha\in \mathbb{R}\}$$ é o conjunto das combinações lineares de $u$. Podemos verificar que $U$ é também um espaço vetorial, que está contido em $V$.

```python
t=np.linspace(0,5,1000) # cria uma lista com 1000 valores entre 0 e 5, igualmente espaçados.
x=[2**s for s in t]
y=[3**s for s in t]

plt.plot(x,y,color='blue')
plt.title('Representação do conjunto U')
plt.show()
```

#**Subspaços vetoriais**



Vimos dois exempos de conjuntos de combinações lineares de elementos de um espaço vetorial, que também são espaços vetoriais.

De forma geral, dado um espaço vetorial $V$ e um conjunto $W\subset V$. Diremos que <font color=blue> **$W$ é subespaço vetorial de $V$** <font color=black> quando $W$ também for um espaço vetorial (com as mesmas definições de operações de $V$).

**Exemplo 1:** Tomando $V=\mathbb{R}^2$ e $u=(1,2)$ então o conjunto 
$$U=\{\alpha u=\alpha(1,2):\, \alpha\in \mathbb{R}\}$$ é também um espaço vetorial, que está contido em $V$.

Então $U$ é subespaço vetorial de $V$.


```python
x=np.linspace(-5,5,1000) # cria uma lista com 1000 valores entre -5 e 5, igualmente espaçados.
y=2*x

plt.plot(x,y,color='blue')
plt.title('Representação do conjunto U')
plt.show()
```

**Exemplo 2:** Seja  $$V = \{(x,y) \in \mathbb{R}^2: x,y > 0 \},$$ o espaço vetorial, com as operações de "adição"
$$\oplus:Q\times Q \rightarrow Q$$
$$(x_1,y_1)\oplus(x_2,y_2)=(x_1x_2,y_1y_2),$$
 e "multiplicação por escalar" dada por 
$$\odot : \mathbb{R} \times Q \rightarrow Q$$
$$c\odot(x,y)=(x^c,y^c).$$ 

Tomando $u=(2,3)\in V$, então o conjunto
$$U=\{\alpha u=(2^{\alpha},3^{\alpha}):\, \alpha\in \mathbb{R}\}$$ é também um espaço vetorial, que está contido em $V$.

Segue que $U$ é um subespaço vetorial de $V$.

```python
t=np.linspace(0,5,1000) # cria uma lista com 1000 valores entre 0 e 5, igualmente espaçados.
x=[2**s for s in t]
y=[3**s for s in t]

plt.plot(x,y,color='blue')
plt.title('Representação do conjunto U')
plt.show()
```

**Exemplo 3 (ou contra-exemplo):** Seja $$V= \{(x,y) \in \mathbb{R}^2: x,y\in\mathbb{R} \}=\mathbb{R}^2$$ que é um espço vetorial, com a soma usual de vetores e produto usual por escalar.

Seja  $$U = \{(x,y) \in \mathbb{R}^2: x,y > 0 \},$$ o espaço vetorial, com as operações de "adição"
$$\oplus:Q\times Q \rightarrow Q$$
$$(x_1,y_1)\oplus(x_2,y_2)=(x_1x_2,y_1y_2),$$
 e "multiplicação por escalar" dada por 
$$\odot : \mathbb{R} \times Q \rightarrow Q$$
$$c\odot(x,y)=(x^c,y^c).$$ 

Já vimos que $U$ é também um espaço vetorial, com as operações $\oplus$ e $\odot$. É verdade também que $U$ está contido em $V$.

Entretanto, $U$ não é subespaço vetorial de $V$, porque $U$ não é espaço vetorial, com as operações de $V$.

Para ver isso, basta observar que, tomando $u=(1,2)$ e $v=(3,4)$ em $U$, então $$u-v=(-2,-2)\not\in U.$$



**Obs:** <font color=blue> Se $V$ for um espaço vetorial e $U$ for um subconjunto de $V$. Então $U$ será um subespaço vetorial de $V$ quando: 



1.   Se $u$ e $v$ forem elementos de $U$ é possível "somar" (com a operação de $V$) esses 
elementos e obter um elemento $w=u+v$ que ainda é um elemento de $U$.
2.   Se $u$ for um elemento de $V$ e $\alpha$ for um elemento do corpo $K$, é possível "multiplicar" (com a operação de $V$) esses elementos e obter um elemento $\alpha u$ que ainda é um elemento de $V$.





#**Espaços gerados por conjunto**

<font color=blue> Quando o espaço vetorial $U$ é uma combinação linear dos elementos do conjunto $C=\{u_{j}:\, j\in D\}\subset V$, em que $V$ é espaço vetorial. Ou seja, quando cada elemento $w$ de $U$ é da forma 
$$w=\alpha_1u_1+\alpha_2u_2+\cdots+\alpha_nu_n,\tag{combinação linear}$$ em que 
$\alpha_1,\,\alpha_2,\,\ldots,\,\alpha_n\in \mathbb{R}$ e $u_1,\,u_2,\,\ldots,\,u_n\in C$, para algum $n\in\mathbb{N}$, dizemos que $U$ é o espaço vetorial gerado pelo conjunto $C$. Podemos usar a notação $U=[C]$ ou $U=G(C)$.

**Exemplo 1:** Tomando $V=\mathbb{R}^2$ e $u=(1,2)$ então o conjunto 
$$U=\{\alpha u=\alpha(1,2):\, \alpha\in \mathbb{R}\}$$ um espaço vetorial gerado pelo conjunto $\{(1,2)\}$ ou pelo conjunto $\{(1,2),\,(0,0),\,(2,4)\}$ por exemplo, que está contido em $V$.

Então $U$ é subespaço vetorial de $V$, gerado pelo conjunto $\{(1,2),\,(0,0),\,(2,4)\}$, por exemplo.



**Exemplo 2:** Seja $V=\mathbb{R}^3$ e seja $D=\{(1,x,y): x,\,y\in \mathbb{R}\}$. Seja $W=[D]$ o espaço gerado por $D$.

Observe que, cada elemento $d=(1,x,y)\in D$ pode ser escrito como 
$$d-(1,1,1)=\alpha((1,3,5)-(1,1,1))+\beta((1,5,7)-(1,1,1)),\tag{$\square$}$$ por exemplo, porque $D$ é um plano, e a expressão anterior é uma de suas [equações vetoriais](http://vectorsandgeometry.wikidot.com/4-4-equacao-do-plano). Essa equação pode ser reescrita como 
$$d=(1,1,1)+\alpha(0,2,4)+\beta(0,4,6).$$

```python
from mpl_toolkits import mplot3d
 
n=20
y = [np.linspace(-2, 2, n) for i in range(n)]   # É preciso fazer uma malha quadriculada n x n.
y=np.array(y)

z = y.copy().T
x = [[1 for i in range(n)] for j in range(n)]
 
fig = plt.figure()
 
ax = plt.axes(projection='3d')
 
ax.plot_surface(x, y, z, cmap='viridis',edgecolor='green')
ax.set_title('Gráfico de plano')
plt.show()
```


Então, o espaço gerado por $D$ é o espaço gerado por elementos $u_i\in D$, da forma 
$$u_i=(1,1,1)+\alpha_i(0,2,4)+\beta_i(0,4,6),$$ ou seja,
$$W=[D]=\left\{\sum_{i=1}^p\theta_iu_i:\,u_i\in D,\,\theta_i\in \mathbb{R},\,p\in\mathbb{N}\right\}.$$ Manipulando a soma da definção de $W$, temos
$$\sum_{i=1}^p\theta_iu_i=a(1,1,1)+b(0,2,3)+c(0,4,6),$$
em que $$a=\sum_{i=1}^p\theta_i,\quad b=\sum_{i=1}^p\theta_i\alpha_i,\quad c=\sum_{i=1}^p\theta_i\beta_i.$$

Observando que, se fixarmos três números $a$, $b$ e $c$, será sempre possível determinar os $\theta_i$, os $\alpha_i$ e os $\beta_i$, da exepressão anterior. Para ver isso, considere as seguintes situações:

1.   Se $a=0$, tome $\theta_1=-\theta_2=1$ e os demais $\theta_i=0$. Depois tome $\alpha_1=b$ e os demais $\alpha_i=0$. Por fim, tome $\beta_1=c$ e os demais $\beta_i=0$
2.   Se $a\neq 0$, tome $\theta_1=1$ e $\theta_2=a-1$  e os demais $\theta_i=0$. Depois tome $\alpha_1=b$ e os demais $\alpha_i=0$. Por fim, tome $\beta_1=c$ e os demais $\beta_i=0$

Isso nos diz que 
$$W=[D]=\{a(1,1,1)+b(0,2,3)+c(0,4,6):\,a,\,b,\,c\in\mathbb{R}\}.$$


Observe ainda que a expressão ($\square$) pode ser escrita como 
$$d=\alpha(1,3,5)+\beta(1,5,7)-(1-\alpha-\beta)(1,1,1),\tag{$\square^*$}$$
o que nos diz que $W$ pode ser gerado por apenas três (dos infinitos) elementos de $D$, ou seja, $$W=\left[\{(1,3,5),(1,5,7),(1,1,1)\}\right].$$ O argumento dessa última igualdade é semelhante ao que usamos antes, e deixamos para o leitor pensar a respeito.

#**Soma de subespaços vetoriais**


Como vimos acima, as combinações lineares de elementos de um subconjunto $C$, qualquer e não vazio, de um espaço vetorial $V$ formam um subespaço vetorial desse espaço, que chamamos de espaço gerado por $C$.

Um conceito útil é o de <font color=blue> soma dos subespaços vetoriais $U$ e $W$ de $V$ <font color=black>, que é definida como
$$U+W=\{u+w:\, u\in U,\,w\in W\}.$$

Como $U$ e $W$ são espaços vetoriais, podemos verificar que 
$U+W$ é o subespaço vetorial de $V$, gerado por $C=U\cup W$.

**Exemplo 1:**

Seja $V=\mathbb{R}^3$, com as operações usuais. 

Seja $C=\{(24,53,12),(1,17,89)\}$ e $W$ o espaço gerado por $C$. 

Seja $D=\{(2,3,4)\}$ e $U$ o espaço gerado por $D$. 

Podemos verificar que $U+ W$ é o espaço gerado por $D\cup C$, que $U$ é uma reta no espaço e que $W$ é um plano no espaço.

Esse espaço é igual a $V$?

Para responder às indagações precisamos pensar se todo vetor $u=(x,y,z)\in V$ é da forma
$$u=\alpha (24,53,12)+\beta(1,17,89)+\delta (2,3,4), $$ 
ou $$\begin{bmatrix}\alpha&\beta&\delta\end{bmatrix}\begin{bmatrix}24&53&12\\1&17&89\\2&3&4\end{bmatrix}=\begin{bmatrix}x&y&z\end{bmatrix}$$
ou 
$$\begin{bmatrix}24&  1&  2\\53& 17&  3\\12& 89 & 4\end{bmatrix}\begin{bmatrix}\alpha\\\beta\\\delta\end{bmatrix}=\begin{bmatrix}x\\y\\z\end{bmatrix}.$$

```python
A=np.array([[24,53,12],[1,17,89],[2,3,4]])
print(f'A=')
print(A)
At=np.transpose(A)
print()
print(f'At=')
print(At)
Ae=At
Ae[1]=Ae[0][0]*Ae[1]-Ae[1][0]*Ae[0]
Ae[2]=Ae[0][0]*Ae[2]-Ae[2][0]*Ae[0]
Ae[2]=Ae[1][1]*Ae[2]-Ae[2][1]*Ae[1]
print(f'Matriz escalonada=')
print(Ae)
```

Usando o processo de eleminação (ou escalonamento), obtemos uma equação equivalente como 
$$\begin{bmatrix} 24& 1& 20\\ 0&355&-34\\  0&  0& 97776\end{bmatrix}\begin{bmatrix}\alpha\\\beta\\\delta\end{bmatrix}=\begin{bmatrix}x\\f(x,y)\\g(x,y,z)\end{bmatrix}.$$ Isso nos diz que todo vetor $v=(x,y,z)\in V$ está no espaço $U+W$, e esses dois espaços são iguais.

**Obs:**

<font color=blue> No caso em que $U\cap W=\{0\}$, a soma de $U$ e $V$ é chamada de soma direta, e denotada por  $$U\oplus W=\{u+w:\, u\in U,\,w\in W,\quad U\cap W=\{0\}\}.$$ 
Nesse caso, definimos o conceito de projeção sobre $U$ ou sobre $W$. Note que, nesse caso, todo elemento $v\in U\oplus W$ é escrito de forma única como $v=u+w$, para $u\in U$ e $w\in W$.

<font color=blue> Por exemplo, definimos a projeção sobre $U$ como 
$${P_U:U\oplus W\Longrightarrow U \atop \quad \quad \quad v\to u}.$$

No caso do exemplo acima, note que a matriz escalonada também nos diz que o vetor no conjunto $D=\{(2,3,4)\}$ não pode ser combinação linear dos vetores de $C$, o que sugere que a soma dos dois espaços $U$ e $W$ é direta.

#**Conjuntos linearmente independentes**

Quando temos um espaço vetorial $V$ e um subconjunto não vazio $C\subset V$ e queremos lidar com o espaço $U$, que é gerado por $C$, nós precisamos pensar em todas as somas (finitas) de elementos de $C$, multiplicados por números de $\mathbb{K}$.

O processo pode ficar mais simples quando $C$ tiver um número finito de elementos apenas, ou quando pudermos lidar com o menor número de elementos possíveis de $C$, ou seja, apenas com elementos de algum subconjunto $D\subset C$ e obtermos o mesmo espaço gerado. 

<font color=blue> Quando não pudermos escolher um subconjuto $D$ que seja menor que $C$, diremos que $C$ é um conjunto linearmente independente (LI).

<font color=blue>  Se houver um subconjunto $D$ que seja menor que $C$ e que ainda gera o mesmo espaço, dizemos que $C$ é linearmente dependente (LD).

**Exemplo 1:** Tomando $V=\mathbb{R}^2$ e o conjunto 
$$U=\{\alpha u=\alpha(1,2)+\beta(2,4)+\delta(0,0):\, \alpha,\beta,\delta\in \mathbb{R}\}$$ que é um subespaço vetorial de $V$, gerado pelo conjunto $C=\{(1,2),\,(0,0),\,(2,4)\}$, por exemplo.

Notamos que não precisamos de todos os elementos de $C$ para gerar $U$. Basta considerarmos o único elemento de $D=\{(1,2)\}$ ou de $D=\{(2,4)\}$, para gerarmos $U$. Então o conjunto $C$ é LD.

**Exemplo 2:**

Seja $V=\mathbb{R}^3$, com as operações usuais. 

Seja $C=\{(24,53,12),(1,17,89),(2,3,4)\}$ e $W$ o espaço gerado por $C$. 

Dos argumentos do penúltimo exemplo, temos que $W=V$ e que $C$ é LI.

#**Base e dimensão de um espaço vetorial**

Dado um espaço vetorial $V$, com as devidas operações, sempre é possível encontrar um subconjunto $C\subset V$ que gera esse espaço $V$.

<font color=blue> Uma base do espaço vetorial $V$ é um conjunto $C$ de $V$, que é $LI$ e que gera o espaço $V$.

<font color=blue> Se $C$ for uma base do espaço vetorial $V$. Dizemos que a dimensão de $V$ é o número de elementos de $C$.

**Obs:** Nesse curso só estamos interessados no caso em que $C$ tem número finito de elementos.

**Exemplo 1** No exemplo anterior temos que $C=\{(24,53,12),(1,17,89),(2,3,4)\}$ é uma base de $V=\mathbb{R}^3$ e que esse espaço vetorial tem dimensão 3.

**Exemplo 2:** Seja $V=\mathbb{R}^3$ e seja $D\{(1,s,t): s,\,t\in \mathbb{R}\}$. Seja $W=[D]$ o espaço gerado por $D$, como no **Exemplo 2 , da Seção: Espaços gerados por conjunto**.

Segue que $D$ é um conjunto $LD$, porque $$W=\left[\{(1,3,5),(1,5,7),(1,1,1)\}\right].$$

Observe ainda que $D_1=\{(1,3,5),(1,5,7),(1,1,1)\}\subset D$ é LI, porque não pode ser reduzido e continuar gerando $W$. 

Na verdade, podemos escrever qualquer vetor $u=(x,y,z)\in V$ como combinação linear dos elementos de $D_1$. Para ver isso basta observar que a expressão 
$$a(1,3,5)+b(1,5,7)+c(1,1,1)=(x,y,z),$$ que é equivalente a 
$$\begin{bmatrix}1&1&1\\3&5&1\\5&7&1\end{bmatrix}\begin{bmatrix}a\\b\\c\end{bmatrix}=\begin{bmatrix}x\\y\\z\end{bmatrix},$$ que é (por escalonamento ou eliminação) equivalente a 
$$\begin{bmatrix}1&1&1\\2&4&0\\4&6&0\end{bmatrix}\begin{bmatrix}a\\b\\c\end{bmatrix}=\begin{bmatrix}x\\y-x\\z-x\end{bmatrix},$$ ou
$$\begin{bmatrix}1&1&1\\2&4&0\\0&-2&0\end{bmatrix}\begin{bmatrix}a\\b\\c\end{bmatrix}=\begin{bmatrix}x\\y-x\\z-x-2(y-x)\end{bmatrix}.$$

Sendo assim, $D_1$ é uma base de $V$, e $W=V$.


<font color=red> **Obs:** Uma observação muito importante sobre o exemplo anterior é que, embora tenhamos visto que $D$ é um plano ( e até escrevemos uma de suas equações), $D$ não é um subespaço de $V$, porque o vetor nulo $u=(0,0,0)$ não está em $D$. Para quem lembra de Geometria Analítica, pode ser que tenha visto que a dimensão de planos é 2, e se o plano contiver o vetor nulo, ele não pode gerar o espaço todo, que tem dimensão 3.

#**Base e dimensão (sistema de coordenadas)**


**Sistema de coordenadas**

<font color=blue> Seja $C=\{u_1,\,u_2,\,\ldots,\,u_n\}$ uma base do espaço vetorial $(V,+,.)$ de dimensão finita $n$. O corpo em questão será $\mathbb{R}$.

Podemos então escrever cada elemento $v\in V$, de forma única, como $$v=\sum_{i=1}^n\alpha_iu_i.$$

<font color=blue> Identificamos então cada vetor $v\in V$ com o vetor $(\alpha_1,\,\alpha_2,\,\ldots,\alpha_n)\in\mathbb{R}^n$. Essa identificação é chamada da sistema de coordenadas (ou representação) de $v$, na base $C$.

**A base canônica de $\mathbb{R}^n$**

Um ponto importante a se destacar é que, quando escrevemos $u=(x_1,\,x_2,\,\ldots,\,x_n)\in \mathbb{R}^n$, e não falamos nada sobre base desse espaço, estamos supondo que 
$$u=(x_1,\,x_2,\,\ldots,\,x_n)=x(1,0,\ldots,0)+x_2(0,1,0,\ldots,0)+\cdots+x_n(0,0,\ldots,0,1)$$ 

<font color=blue> Isso nos diz que $$C=\{(1,0,\ldots,0),\,(0,1,0,\ldots,0),\,\ldots,\,(0,0,\ldots,0,1)\}$$ é uma base de $\mathbb{R}^n$, que chamamos de base canônica.

**Exemplo:** Seja $V=\mathbb{R}^3$ e seja $D\{(1,s,t): s,\,t\in \mathbb{R}\}$. Seja $W=[D]$ o espaço gerado por $D$, como no **Exemplo 2 , da Seção: Espaços gerados por conjunto**.

Segue já vimos que $D$ é um conjunto $LD$, porque $$W=\left[\{(1,3,5),(1,5,7),(1,1,1)\}\right].$$

Observe ainda que $D_1=\{(1,3,5),(1,5,7),(1,1,1)\}\subset D$ é LI, porque não pode ser reduzido e continuar gerando $W$. 

Já vimos também que, podemos escrever qualquer vetor $u=(x,y,z)\in V$ como combinação linear dos elementos de $D_1$. Para ver isso basta observar que a expressão 
$$a(1,3,5)+b(1,5,7)+c(1,1,1)=(x,y,z),$$ que é equivalente a 
$$\begin{bmatrix}1&1&1\\2&4&0\\0&-2&0\end{bmatrix}\begin{bmatrix}a\\b\\c\end{bmatrix}=\begin{bmatrix}x\\y-x\\z-x-2(y-x)\end{bmatrix}.$$

Sendo assim, $D_1$ é uma base de $V$, e $W=V$.

Nesse caso, <font color=green> podemos escrever cada elemento de $u=(x,y,z)\in V$, na base canônica, <font color=black> usando a representação dela nessa base $D_1$, usando o sistema linear acima. Então, por exemplo, tomando $u=(1,2,3)$ teremos que $u=(a,b,c)_{D_1}$, representado na base $D_1$, em que $a$, $b$ e $c$ são determinados pelo sistema:
$$a(1,3,5)+b(1,5,7)+c(1,1,1)=(1,2,3),$$ que é equivalente a
$$\begin{bmatrix}1&1&1\\2&4&0\\0&-2&0\end{bmatrix}\begin{bmatrix}a\\b\\c\end{bmatrix}=\begin{bmatrix}1\\1\\0\end{bmatrix},$$ que nos diz que 
$b=0$, $a=1/2$ e $c=1/2$.



```python
D1e=[[1,3,5],[1,5,7],[1,1,1],[1,2,3]]
D1e=np.array(D1e)
D1e=np.transpose(D1e)
print(D1e)
print()
D1e[1]=D1e[1]-D1e[0]
D1e[2]=D1e[2]-D1e[0]
print(D1e)
print()
D1e[2]=D1e[2]-2*D1e[1]
print(D1e)
```

<font color=blue> Então o vetor $u=(1,2,3)$, na base canônica de $\mathbb{R}^3$, é escrito como $u=(0.5,0,0.5)$ na base $D_1$.

**Exercício 28 do livro do Steinbruch (Seção 2.10):**

Sejam $p_1(x)=x^2-2x+1$, $p_2(x)=x+2$ e $p_3(x)=2x^2-x$.

a) Esccreva $p(x)=5x^2-5x+7$ como combinação linear de $p_1(x)$, $p_2(x)$ e $p_3(x)$.

b) Escrever $p(x)$ como combinação linear de $p_1(x)$ e $p_2(x)$.

c)...

d) Escrever $p_1(x)$ como combinação linear de $p_2(x)$ e $p_3(x)$.



*Solução:* 

a) Precisamos determinar o vetor $u=(a,b,c)$ tal que 
$$p(x)=ap_1(x)+bp_2(x)+cp_3(x).$$ Isso quer dizer que 

$$a(x^2-2x+1)+b(x+2)+c(2x^2-x)=5x^2-5x+7$$ para qualquer $x$ escolhido. Reescrevendo a expressão obtemos 
$$(a+2c)x^2+(-2a+b-c)x+(a+2b)=5x^2-5x+7$$ o que implica que 
$$\begin{bmatrix}1&0&2\\-2&1&-1\\1&2&0\end{bmatrix}\begin{bmatrix}a\\b\\c\end{bmatrix}=\begin{bmatrix}5\\-5\\7\end{bmatrix}.$$

```python
D1e=np.array([[1,-2,1],[0,1,2],[2,-1,0],[5,-5,7]])
D1e=np.array(D1e)
D1e=np.transpose(D1e)
print(D1e)
print()
D1e[2]=D1e[2]-D1e[0]
D1e[1]=D1e[1]+2*D1e[0]
print(D1e)
print()
D1e[2]=D1e[2]-2*D1e[1]
print(D1e)
print()
```

Esse último sistema é quivalente ao sistema escalonado, <font color=blue>(obtido como combinação linear das linhas da matriz extendida do problema):

$$\begin{bmatrix}1 & 0 &  2\\0 & 1&  3\\ 0&  0 &-8 \end{bmatrix}\begin{bmatrix}a\\b\\c\end{bmatrix}=\begin{bmatrix}5\\5\\-8\end{bmatrix}.$$

Isso nos diz que $c=1$, $b=2$ e $a=3$.

<font color=green> Note que essa eliminação nos diz que o conjunto $\{p_1(x),\,p_2(x),\,p_3(x)\}$ é linearmente independende $LI$. Diz também que o item b) e d) desse exercício gera um sistema linear impossível.
