Isso quer dizer, por exemplo,que $F$ consome metada do que produz, metade do que $C$ produz e um terço do que $M$ produz, equanto fornece um quarto do que produz para $M$ e um quarto para $C$.

Isso pode ser escrito como as equações:$$\begin{cases}\frac{1}{4}x_1+\frac{1}{4}x_1-\frac{1}{3}x_2-\frac{1}{2}x_3&=&0\qquad \text{em $F$}\\\\\frac{1}{3}x_2+\frac{1}{3}x_2-\frac{1}{4}x_1-\frac{1}{4}x_3&=&0\qquad \text{em $M$}\\\\\frac{1}{2}x_3+\frac{1}{4}x_3-\frac{1}{3}x_2-\frac{1}{4}x_1&=&0\qquad \text{em $C$} \end{cases},$$ que podem ser reescritas como

$$\begin{bmatrix}\frac{1}{2}&-\frac{1}{3}&-\frac{1}{2}\\\\-\frac{1}{4}&\frac{2}{3}&-\frac{1}{4}\\\\-\frac{1}{4}&-\frac{1}{3}&\frac{3}{4}\end{bmatrix}\begin{bmatrix}x_1\\\\x_2\\\\x_3\end{bmatrix}=\begin{bmatrix}0\\\\0\\\\0\end{bmatrix},$$
considerando a [notação e multiplicação usual de matrizes](https://pt.wikipedia.org/wiki/Sistema_de_equa%C3%A7%C3%B5es_lineares), o que já usa conceitos mais avançados de matemática, digamos assim.

Segue que, podemos usar as igualdades do sistema linear para somar equações e obter outras equações, por exemplo:
$$\left(\frac{1}{2}x_1-\frac{1}{3}x_2-\frac{1}{2}x_3\right)+\left(-\frac{1}{4}x_1+\frac{2}{3}x_2-\frac{1}{4}x_3\right)+\left(-\frac{1}{4}x_1-\frac{1}{3}x_2+\frac{3}{4}x_3\right)=0+0+0=0,$$ ou $$\left(\frac{1}{2}x_1-\frac{1}{3}x_2-\frac{1}{2}x_3\right)+2\times \left(-\frac{1}{4}x_1+\frac{2}{3}x_2-\frac{1}{4}x_3\right)=x_2-x_3=0+2\times 0=0$$  e $$\left(\frac{1}{2}x_1-\frac{1}{3}x_2-\frac{1}{2}x_3\right)+2\times \left(-\frac{1}{4}x_1-\frac{1}{3}x_2+\frac{3}{4}x_3\right)=-x_2+x_3=0+2\times 0=0.$$

As duas últimas somas de equações produzem a igualdade $$x_2-x_3=0\Longrightarrow x_2=x_3.$$

Usando, por exemplo, a equação que obtemos em $F$, temos que
$$0=\frac{1}{2}x_1-\frac{1}{3}x_2-\frac{1}{2}x_3=\frac{1}{2}x_1-\frac{1}{3}x_2-\frac{1}{2}\color{blue}{x_2}=\frac{1}{2}x_1-\frac{5}{6}x_2\Longrightarrow x_1=\frac{5}{3}x_2.$$

Tomando $x_2=x_3=3$, por exemplo, temos $x_1=5$ como uma possíbilidade de solução.

$$x_1\begin{bmatrix}\frac{1}{2}\\\\-\frac{1}{4}\\\\-\frac{1}{4}\end{bmatrix}+x_2\begin{bmatrix}-\frac{1}{3}\\\\\frac{2}{3}\\\\-\frac{1}{3}\end{bmatrix}+x_3\begin{bmatrix}-\frac{1}{2}\\\\-\frac{1}{4}\\\\\frac{3}{4}\end{bmatrix}=\begin{bmatrix}0\\\\0\\\\0\end{bmatrix}.$$

**Obs:** Os produtos que aparecem na expessão $(\circ)$ são chamados de produtos escalares; e podem ser vistos como produtos de uma matriz linha por uma matriz matriz coluna, de mesma dimensão. Por exemplo, $$(u+2w)\begin{bmatrix}x_1\\\\x_2\\\\x_3\end{bmatrix}=\begin{bmatrix}0&-1&1\end{bmatrix}\begin{bmatrix}x_1\\\\x_2\\\\x_3\end{bmatrix}=0\times x_1-1\times x_2+1\times x_3.$$



De modo geral, se $k=(k_1,\,k_2,\,\ldots,\,k_n)$ e $l=(l_1,\,l_2,\,\ldots,\,l_n)$ são dois vetores (como em Geometria Analítica) o produto escalar entre $k$ e $k$ é dado por $$k\cdot l=\sum_{i=1}^nk_i\times l_i.$$

**Obs:** O produto escalar entre $k$ e $l$ pode ser obtido da composição de duas outras operações, [uma binária](https://pt.wikipedia.org/wiki/Opera%C3%A7%C3%A3o_bin%C3%A1ria) e outra função que soma as coordenadas dos vetores.

Nesse caso, a operação binária é um caso particular do [produto de Hadamard](https://en.wikipedia.org/wiki/Hadamard_product_(matrices)) $$k\odot l=(k_1\times l_1,\,k_2\times l_2,\,\ldots,k_n\times l_n).$$

Por exemplo, para $k=(2,3,5)$ e $l=(1,4,7)$ temos $$k\odot l=(2\times 1,3\times 4,5\times 7)=(2,12,35)$$ e o produto escalar
$$k\cdot l=2\times 1+3\times 4+5\times 7=2+12+35=49.$$

Com a calculadora daqui:

```python
k,l=np.array([2,3,5]),np.array([1,4,7])
print('\n k=',k,'\n l=',l)
```

```python
np.dot(k,l)               # np.dot() é a função que calcula produto escalar.
```

```python
k*l                       # * calcula o produto de Hadamard.
```

```python
sum(k*l)                  # soma das coordenadas de k*l.
```

```python
k@l                       # No caso de matrizes linha o @ coincide como o np.dot().
```

#**Obs: Sobre listas e soma de listas aqui no Python**

Note que, se não usar o pacote [numpy](https://numpy.org/doc/stable/user/absolute_beginners.html), quando somar, por exemplo, as listas, $$r=[2,3,4],\qquad s=[3,4,5],$$ vai obter uma lista maior, ou seja, o Python não vai ver a lista como um vetor usual.

```python
r=[2,3,4]
s=[3,4,5]
w=r+s
print('w=r+s=',w)
```

Note ainda que listas podem ter entradas que não são números.

Por exemplo,

```python
k=[[1,2,'casa'],'queijo',r,s,[r,s,w]]                         # Listas com números, palavras e outras listas como elementos.
m=['bola',['bola','bolo','bola quadrada'],7]
print('\n k=',k,',\n m=',m,',\n k+m=',k+m)
```

**Obs:** Pode ocorrer de usarmos listas como representação de conjuntos. Por exemplo, os conjuntos $$A=\{1,2,3\},\qquad B=\{a,b,c,6\}$$ podem ser representados como uma lista com os três números $1$, $2$ e $3$ (ou $[1,2,3]$) e uma lista com letras e um número $a$, $b$, $c$ e $6$ (ou $[a,b,c,6]$).

```python
A=[1,2,3,2,1]       # Lista
B=['a','b','c',6]   # Lista
A=set(A)         # Muda para conjunto.
B=set(B)         # Conjunto.
A,B              # Par ordenado de conjuntos.
```

E pode fazer a união de conjuntos como 'soma'.

```python
A=[1,2,3,2,1]       # Lista
B=['a','b','c',6]   # Lista
set(A+B)
```
