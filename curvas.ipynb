**Curvas paramétricas**

Este pequeno texto é  para relembrar algumas propriedades das curvas que podem ser expressas como $c(t)=(x(t),y(t))$, em que $t$ é um número real, por exemplo.

Considere as funções $f(s)=s^2+2s+1$ e $g(s)=s^2-3s+2$.

```python
def f(s):              #Definindo as funções em Python.
  return s**2+2*s+1

def g(s):
  return s**2-3*s+2
```

Estamos bastante acostumados, em cursos de exatas, a fezer gráficos de funções como $f(s)$ e $g(s)$.

O gráfico de $f(s)$, por exemplo, é o conjunto dos pontos do plano cartesiano da forma $(s,f(s))$.

Lembramos como fazer isso em Python:

```python
import numpy as np, matplotlib.pyplot as plt

x=np.linspace(-2,4,100)
y=x
plt.plot(x,f(x))
plt.title('Gráfico de f(s)')
plt.xlabel('s')
plt.ylabel('f(s)')
plt.show()


plt.plot(x,g(x))
plt.xlabel('s')
plt.ylabel('g(s)')
plt.title('Gráfico de g(s)')
plt.show()
```

Podemos também considerar o conjunto dos pontos do plano cartesiano da forma $(f(s),g(s))$. Isso é o que chamamos de curva paramétrica.

```python
plt.plot(f(x),g(x))
plt.xlabel('f(s)')
plt.ylabel('g(s)')
plt.title('Curva paramétrica (f(s),g(s))')
plt.show()
```

Nesse caso, a curva  $c(s)=(f(s),g(s))$ é uma função $c:\mathbb{R}\to \mathbb{R}\times \mathbb{R}$, e a imagem anterior não é um gráfico de $c(s)$ e sim a representação do conjunto imagem da função $c$.

Vejamos mais um exemplo em dimensão maior.

Neste caso, tomamos a curva $C(s)=(cos(s^4+2s^2+1),(s^2+1)\cos(s),(s^2+1)sen(s))=(k(s),p(s),w(s))$.

```python
ax = plt.figure().add_subplot(projection='3d')

# Prepare arrays x, y, z
z = np.linspace(-2, 2, 200)
r = z**2 + 1
k = r * np.sin(z)
p = r * np.cos(z)
w=np.cos(z**4+2*z**2+1)

ax.plot(k, p, w, label='Curva C(s)')
ax.legend()

plt.show()
```

**A derivada de curvas paramétricas**

Quando temos uma curva paramátircas $c(s)$, em que $s$ é um número real, definimos a derivada $c'(s)$, caso exista, como $$c'(s)=\lim_{h\to 0} \frac{c(t+h)-c(t)}{h}.$$

Em particular, se $c(s)$ tiver coordenadas deriváveis, então a derivada $c'(s)$ terá a derivada de cada coordenada de $c(s)$ como respectiva coordenada.



Como exemplo, temos que, se $f(s)=s^2+2s+1$, $g(s)=s^2-3s+2$ e $c(s)=(f(s),g(s)$. Então $c'(s)=(f'(s),g'(s))=(2s+2,2s-3)$.

```python
def df(s):
  return 2*s+2

def dg(s):
  return 2*s-3
```

Cosideramos os pontos $s\in [-1,0,1,2,3]$ para representar $c'(s)$ no gráfico.

```python
plt.plot(f(x),g(x),label='c(s)')
plt.xlabel('f(s)')
plt.ylabel('g(s)')
plt.title('Curva paramétrica c(s)=(f(s),g(s))')
for s in [-1,0,1,2,3]:
  plt.quiver(f(s),g(s),df(s),dg(s),scale_units='xy', angles='xy', scale=1, color='green',width=0.005)
plt.ylim(-2,13)
plt.xlim(-2,26)
plt.legend()
plt.show()
```

**Obs:** A derivada de uma curva $c(s)$ nos dá vetores $c'(s)$ tangentes à curva, no ponto $c(s)$.

**A integral de uma curva peramétrica**

Como a derivada $c'(s)$ de uma curva $c(s)$ é obtida por meio das derivadas das coordenadas dessa curva, podemos definir as primitivas de $c(s)$ usando primitivas de cada uma das suas coordenadas.

Como exemplo, se $r(s)=(2s+2,2s-3)$, então uma primitiva dessa curva é dada por
$$R(s)=\int c(s)\,ds=(s^2+2s+k_1,s^2-3s+k_2),$$ para cada escolha de $k_1,\,k_2\in\mathbb{R}$.

**Obs:** Se soubermos um valor para, digamos $R(s_0)$, então as constantes $k_1,\,k_2$ podem ser escolhidas.

No caso da curva anterior, se tomarmos $R(0)=(1,2)$ obtemos que $T(s)=c(s)=(s^2+2s+1,s^2-3s+2)$.

**Equações diferencias ordinárias**

Dados os comentários deste texto, sobre curvas $c(s)\in \mathbb{R}^n$, para $s\in (a,b)\subset \mathbb{R}$, uma equação diferencial ordinária é uma expressão da forma
$c'(s)=F(s,c(s))$, em que a função $F(s,u)$, para $u\in \mathbb{R}^n$, é conhecida e a função $c(s)$ é desconhecida.

Quando conhecemos $c(s_0)=c_0$ e queremos encontrar $c(s)$, chamamos a expressão
$$\begin{cases}c'(s)&=&F(s,c(s))\\\\c(s_0)&=&c_0\end{cases},$$
de [problema de valor inicial](https://pt.wikipedia.org/wiki/Problema_de_valor_inicial#:~:text=Um%20problema%20de%20valor%20inicial%20%28%20P.V.I.%29%20%C3%A9,que%20%C3%A9%20solu%C3%A7%C3%A3o%20da%20equa%C3%A7%C3%A3o%20diferencial%20e%20satisfaz.).

Como exemplo, considere a curva $c(t)=(x(t),y(t))$, tal que $$\begin{cases}x'(t)&=&-y(t)\\\\y'(t)&=&x(t)\\\\x(0)&=&1\\\\y(0)&=&0\end{cases}.$$

Observe que a curva $c(t)=(\cos(t),sen(t))$ é a função definida pelo problema de valor inicial dado, porque $c'(t)=(-sen(t),cos(t))$ e $c(0)=(1,0)$.

Pode-se mostrar que esta é a única possibilidade, para todo $t\in\mathbb{R}$.
