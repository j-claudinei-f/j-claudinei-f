A ideia deste texto é fazer um exercício visual do teorema de Thales, partindo da validade de resultado demonstrado em aula (Proposição 2.35 do livro de Geometria do ProfMat).

Considere três pontos $A$, $B$ e $C$ do plano (aqui é preciso escolher coordenadas desses pontos). Considere também um ponto $D$, do segmento $AB$.

<font color=blue> Você pode alterar os valores numéricos no código abaixo:

```python
A=(2,5)
B=(-1,3)
C=(3,1)
s=2**(1/2)/2
print('s=',s)
```

Na simulação o ponto $D$ é obtido da relação $D=A+s(B-A)$, ou seja, $\overline{AD}=s$, na escala $\overline{AB}=1$. Então $\overline{DB}=1-s$.

Escolhi $s=\frac{\sqrt{2}}{2}$, mas você pode trocar por outro valor entre 0 e 1.

<font color=red> Não mexa na parte do código abaixo:

```python
import numpy as np

A=np.array(A)
B=np.array(B)
C=np.array(C)

D=A+s*(B-A)
E=A+s*(C-A)

x1=[A[0],B[0],D[0]]
y1=[A[1],B[1],D[1]]

x2=[A[0],C[0],E[0]]
y2=[A[1],C[1],E[1]]

import matplotlib.pyplot as plt

def gbase():
  plt.plot(x1,y1,color='blue')
  plt.plot(x1,y1,'o',color='orange')

  plt.plot(x2,y2,color='blue')
  plt.plot(x2,y2,'o',color='orange')

  plt.plot([D[0],E[0]],[D[1],E[1]],color='red')
  plt.plot([B[0],C[0]],[B[1],C[1]],color='red')

  L=['A','B','C','D','E']
  K=[A,B,C,D,E]
  for j in range(0,5):
    plt.text(K[j][0], K[j][1],L[j], fontsize=12)

gbase()
```

A ideia agora é dividir o segmento $AD$ em $n$ partes congruentes e depois ir acrescentando segmentos de tamanho $\overline{AD}/n$ até passar do ponto $B$.

Fica implícito que $DE$ é paralelo a $BC$.

<font color=red> Não mexa no código daqui para frente:

```python
n=4
```

```python
n=4
t=1/n
D1=A+t*(D-A)
E1=A+t*(E-A)
u=D1-A
v=E1-A

D2=D1
E2=E1


plt.plot([D2[0],E2[0]],[D2[1],E2[1]],color='green')

m=1
while D2[1]>B[1]:
  D2=D2+u
  E2=E2+v
  plt.plot([D2[0],E2[0]],[D2[1],E2[1]],color='green')
  m=m+1


gbase()
plt.show()

print('(n,m)=(',n,',',m-n,')')
print('n/m=',n/(m-n))
```

A ideia é observar que existem $n$ e $m$ números naturais, tais que $$\frac{\overline{AD}}{\overline{DB}}\approx\frac{n}{m}\approx \frac{\overline{AE}}{\overline{EC}},$$ com a precisão que você desejar.
