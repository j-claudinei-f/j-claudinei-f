#**Simulção Plano de Carreira Docentes Federais**

#**Proposta Andes de reajuste de 22.71% em três anos, com ideia de base ser salário de 20 horas**

Veja planilha abaixo.

```python
pi=3839.21*(1+22.71/100)

p=[[1,pi,2*pi,3.1*pi]]

for i in range(2,14):
 pi=pi*(5/100+1)
 p.append([i,pi,2*pi,3.1*pi])
```

```python
import pandas as pd
Indice=['Probatório 1','Probatório 2','Assistente 1', 'Assistente 2','Adjunto 1','Adjunto 2','Adjunto 3','Adjunto 4','Associado 1','Associado 2','Associado 3','Associado 4','Titular']
df=pd.DataFrame(p,index=Indice,columns=['nível','20 horas','40 horas', 'DE'])
import matplotlib. pyplot as plt

#initialize figure
fig = plt. figure (figsize = (8, .2))
ax = fig. add_subplot (111)

#create table
ax. table (cellText = df. values , rowLabels = df. index ,
         colLabels = df. columns )

#add title to table
ax. set_title (' Propsta carreira - steps de 5%, como peso 1, 2 e 3.1, partindo de 20 horas')

#turn axes off
ax.axis('off')
plt.show()
```

#**Proposta Governos de reajuste de 9% em 2025 e 3.5 em 2026, com alteração de carreira**

Veja planilha abaixo.

```python
import numpy as np
salarioatual=np.array([12191.60,12862.13,13376.61,13911.69,14468.14,18085.19,8748.19+10060.41,19560.93,20343.37,22377.72])

Salario2025=salarioatual*(1+9/100)
Salario2026=salarioatual*(1+9/100)*(1+3.5/100)
Salario2025novapropsta=[salarioatual[0]*(1+9/100)] # Probatório
p=salarioatual[1]*(1+9/100)
Salario2025novapropsta.append(p) #Adjunto 1
p=p*(1+4.5/100)
Salario2025novapropsta.append(p) #Adjunto 2
p=p*(1+4.5/100)
Salario2025novapropsta.append(p) #Adjunto 3
p=p*(1+4.5/100)
Salario2025novapropsta.append(p) #Adjunto 4
p=p*(1+23.5/100)
Salario2025novapropsta.append(p) #Associado 1
p=p*(1+4.5/100)
Salario2025novapropsta.append(p) #Associado 2
p=p*(1+4.5/100)
Salario2025novapropsta.append(p) #Associado 3
p=p*(1+4.5/100)
Salario2025novapropsta.append(p) #Associado 4
Salario2025novapropsta.append(salarioatual[9]*(1+9/100)) #Titular

Salario2026novapropsta=salarioatual*np.array([(1+9/100)*(1+3.5/100),1+13.3/100,1+14.3/100,1+15.5/100,1+16.7/100,1+14.3/100,1+15.4/100,1+16.5/100,1+17.6/100,1+17.61/100])
Proposta=[salarioatual,Salario2025,Salario2026,Salario2025novapropsta,Salario2026novapropsta]
```

```python
df=pd.DataFrame(Proposta,index=[2024,2025,2026,'2025-novo','2026-novo'],columns=['Probatório','Adjunto 1','Adjunto 2','Adjunto 3','Adjunto 4','Associado 1','Associado 2','Associado 3','Associado 4','Titular'])
#initialize figure
fig = plt. figure (figsize = (20, .5))
ax = fig. add_subplot (111)

#create table
ax. table (cellText = df. values , rowLabels = df. index ,
         colLabels = df. columns )

#add title to table
ax. set_title (' Proposta carreira - com reajuste de 9% e 3.5%, com e sem mudança de steps')

#turn axes off
ax.axis('off')
plt.show()
```
